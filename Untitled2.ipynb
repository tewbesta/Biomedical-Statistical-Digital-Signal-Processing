```python
import matplotlib.pyplot as plt
import numpy as np
Fs = 10000
t=1
f = (100*t + ((400/3)*(t**3)))
sample = 10000
x = np.arange(0,sample)
y = np.cos(f * x/ Fs)
t=np.arange(0,1,1/Fs)
plt.plot(t, y)
plt.xlabel('sample(n)')
plt.ylabel('Amplitude')
noise= np.random.normal(0, 0.5, 10000)
plt.plot(t,noise)
plt.show()
```

```python

from scipy import signal
Nfft=sample
fs=Fs
cutoff=1000
h = signal.firwin(sample, cutoff, pass_zero='lowpass', fs=fs)
H = np.fft.fft(h, n=Nfft)
f = np.arange(0, Nfft) 
H = H[f < fs/2]
S_true = np.abs(H) ** 2
S_true /= S_true.sum()
w = signal.windows.hann(sample)
X = np.fft.fft(y * w[:, None], Nfft, axis=0)
f = np.arange(0, Nfft) * fs/Nfft
X = X[f < fs/2]
f = f[f < fs/2]
S_periodogram = (np.abs(X) ** 2).mean(axis=1) / Nfft
S_periodogram /= S_periodogram.sum()
f = f[f < fs/2]
plt.plot(f, 10 * np.log10(S_true))
plt.plot(f, 10 * np.log10(S_periodogram))
plt.xlim((0, fs/2))
plt.xlabel('Frequency (Hz)')
plt.ylabel('S(f)')
plt.legend(('True', 'Periodogram'))
```
