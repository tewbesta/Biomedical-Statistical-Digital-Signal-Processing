```python
#1d
def ar1(a,Nexamples,Nsamps):
    xt=np.zeros((Nexamples,Nsamps))
    for k in range(Nexamples):
      x=np.zeros((Nsamps))
      w = np.random.randn(N) 
      x[0]=w[0]
      for n in range(1,N):
       # Generate 2000 sample white noise
        x[n]=a*x[n-1]+w[n]
      std=np.std(x)
      xt[k,:]=x/std
      
      pl.plot(xt[k,1000:]) 
    avgvar=np.mean(np.var(xt[:],axis=1))
    print(avgvar)
    return xt,avgvar
```

```python
#1e
aa,var=ar1(0.999,10,2000)
print(var)
pl.show()
ar1(0.99,10,2000)
```
